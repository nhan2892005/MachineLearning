# Random number does NOT mean a different number every time. Random means something that can not be predicted logically.

```python
from numpy import random
```

## generate random number

Numpy provided the random module to work with random numbers.

```python
x = random.randint(100)

print(x)
```

## generate random float

will return number between 0 and 1

```python
x = random.rand()

print(x)
```

```python
# array random
arr = random.rand(3, 5) * random.randint(100, size=(5))
arr
```

```python
x = random.choice(arr.reshape(-1), size = (4,2))
x
```

```python
x = random.choice([3, 5, 7, 9], p=[0.3, 0.2, 0.5, 0.0] ,size=(3, 5))
x
```

## random permutation

```python
random.shuffle(arr)
arr
```

```python
x = random.permutation(arr)
x
```

## visualize distribution with Seaborn

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
sns.displot(random.rand(1000), kde = True)

plt.show()
```

## normal distribution

```python
data = random.normal(loc=10, scale=6, size=100)

sns.displot(data, kde = True)

plt.show()
```

## binomial distribution

```python
bino = random.binomial(n=10, p=0.5, size=1000)

sns.displot(bino, kde = True)

plt.show()
```
